```python
Xf,Xt,Xv,Yf,Yt,Yv,Yf_pred,Yt_pred,Yv_pred,err = [],[],[],[],[],[],[],[],[],[]
Lambda = [1]
Wf,Wt,Wv = np.zeros((10,1)),np.zeros((10,1)),np.zeros((10,4))
P49_ref = 0.353
#### Estimate Sensitivities
def sensitivity(start):
    lmbda = 0.85
    
    dVolt = [x-y for x,y in zip(Voltages.iloc[-1].tolist()[1:],Voltages.iloc[-10].tolist()[1:])]
    dFreq = [sum(Frequencies.iloc[-1].tolist()[1:])/5.0-sum(Frequencies.iloc[-10].tolist()[1:])/5.0]
    dP49 = [Active_power.iloc[-1].tolist()[-1]-Active_power.iloc[-10].tolist()[-1]]
    Yf.append(dFreq)
    Yt.append(dP49)
    Yv.append(dVolt)
    
    Xf.append([P1_array[-1][1],P2_array[-1][1],P3_array[-1][1],P6_array[-1][1],P8_array[-1][1]])
    Xf[-1].extend([Q1_array[-1][1],Q2_array[-1][1],Q3_array[-1][1],Q6_array[-1][1],Q8_array[-1][1]])
    Xt.append([P1_array[-1][1],P2_array[-1][1],P3_array[-1][1],P6_array[-1][1],P8_array[-1][1]])
    Xt[-1].extend([Q1_array[-1][1],Q2_array[-1][1],Q3_array[-1][1],Q6_array[-1][1],Q8_array[-1][1]])    
    Xv.append([P1_array[-1][1],P2_array[-1][1],P3_array[-1][1],P6_array[-1][1],P8_array[-1][1]])
    Xv[-1].extend([Q1_array[-1][1],Q2_array[-1][1],Q3_array[-1][1],Q6_array[-1][1],Q8_array[-1][1]])
    
#     Xv[-1].append(sum([x-y for x,y in zip(Frequencies.iloc[-1].tolist()[1:],Frequencies.iloc[start].tolist()[1:])])/5)
#     Xv[-1].extend([x-y for x,y in zip(Active_power.iloc[-1].tolist()[1:],Active_power.iloc[start].tolist()[1:])])
#     Xv[-1].extend([x-y for x,y in zip(Reactive_power.iloc[-1].tolist()[1:],Reactive_power.iloc[start].tolist()[1:])])
#     Xv[-1].extend([x-y for x,y in zip(Sw_loads.iloc[-1].tolist()[1:],Sw_loads.iloc[-3].tolist()[1:])])
        
    Yf_pred.append(list(np.dot(np.array(Xf[-1]),Wf)))
    Yt_pred.append(list(np.dot(np.array(Xt[-1]),Wt)))
    Yv_pred.append(list(np.dot(np.array(Xv[-1]),Wv)))
    X,Y = np.array(Xf),np.array(Yf)
    L = Lambda[::-1]
    Sf = multi_dot([np.linalg.pinv(np.dot(X.T*np.array(L),X)),X.T*np.array(L),Y])
    X,Y = np.array(Xt),np.array(Yt)
    St = multi_dot([np.linalg.pinv(np.dot(X.T*np.array(L),X)),X.T*np.array(L),Y])
    X,Y = np.array(Xv),np.array(Yv)
    Sv = multi_dot([np.linalg.pinv(np.dot(X.T*np.array(L),X)),X.T*np.array(L),Y])
    Lambda.append(Lambda[-1]*lmbda)
    
    err.append([100*abs(x-y)/max(abs(x),1e-4) for x,y in zip(Yf[-1],Yf_pred[-1])]+\
               [100*abs(x-y)/max(abs(x),1e-4) for x,y in zip(Yt[-1],Yt_pred[-1])]+\
               [100*abs(x-y)/max(abs(x),1e-4) for x,y in zip(Yv[-1],Yv_pred[-1])])
    print(f"error is {err[-1]}")
    
    return Sf,St,Sv
#### Solve optimization to compute secondary control input adjustment
def control_perturb(k):
    ave_freq = sum(Frequencies.iloc[-1].tolist()[1:])/5.0
    P49 = Active_power.iloc[-1].tolist()[-1]
    ave_volt = sum(Voltages.iloc[-1].tolist()[1:])/3.0
    y = np.array([np.clip(60-ave_freq,-0.02,0.02)]+[np.clip(P49_ref-P49,-0.02,0.02)]+\
                 [np.clip(1-Voltages.iloc[-1].tolist()[1],-0.03,0.03)]+\
                 [np.clip(0.92-Voltages.iloc[-1].tolist()[2],-0.03,0.03)]+\
                 [np.clip(0.93-Voltages.iloc[-1].tolist()[3],-0.03,0.03)])
    A = np.r_[Wf.T,Wt.T,Wv.T]
    I = np.identity(10)
    Diag = np.diag([1,1,1,1,1])
    rho = 5e-4
    np.random.seed(k)
    if k<80:
        u = [np.clip((60-ave_freq),-0.05,0.05)]*5+[np.clip(5*(1-ave_volt),-0.2,0.2)]*5
        a = [0.25,0.5,0.75,1]    
        d = [a[i] for i in np.random.randint(4,size=(10,))]
        return [x*y for x,y in zip(u,d)] 
    else:        
        P = A.T @ A
        q = -2*A.T@y
        G = np.r_[I,-I]
        h = np.ones(20)*0.06-np.random.uniform(0,0.03,size=(20,))
        x = cp.Variable(10)
        prob = cp.Problem(cp.Minimize(cp.quad_form(x, P) + q.T @ x),[G @ x <= h])
        prob.solve()
#         u = multi_dot([pinv(multi_dot([A.T,Diag,A])+rho*I),A.T,Diag,y])
        return list(x.value)
```

```python
folder = "Data/Data 3/"
P1 = pd.read_csv(folder+"P1.csv",index_col = False,header=None)
P2 = pd.read_csv(folder+"P2.csv",index_col = False,header=None)
P3 = pd.read_csv(folder+"P3.csv",index_col = False,header=None)
P8 = pd.read_csv(folder+"P8.csv",index_col = False,header=None)
Q1 = pd.read_csv(folder+"Q1.csv",index_col = False,header=None)
Q2 = pd.read_csv(folder+"Q2.csv",index_col = False,header=None)
Q3 = pd.read_csv(folder+"Q3.csv",index_col = False,header=None)
Q8 = pd.read_csv(folder+"Q8.csv",index_col = False,header=None)
N = 400
P1_array = [[0,0] for _ in range(N+1)]
P2_array = [[0,0] for _ in range(N+1)]
P3_array = [[0,0] for _ in range(N+1)]
P8_array = [[0,0] for _ in range(N+1)]
Q1_array = [[0,0] for _ in range(N+1)]
Q2_array = [[0,0] for _ in range(N+1)]
Q3_array = [[0,0] for _ in range(N+1)]
Q8_array = [[0,0] for _ in range(N+1)]
for k in range(N+1):
    P1_array[k] = P1.iloc[k].tolist()
    P2_array[k] = P2.iloc[k].tolist()
    P3_array[k] = P3.iloc[k].tolist()
    P8_array[k] = P8.iloc[k].tolist()
    Q1_array[k] = Q1.iloc[k].tolist()
    Q2_array[k] = Q2.iloc[k].tolist()
    Q3_array[k] = Q3.iloc[k].tolist()
    Q8_array[k] = Q8.iloc[k].tolist()
P1_array = np.array(P1_array)   
P2_array = np.array(P2_array)  
P3_array = np.array(P3_array)  
P8_array = np.array(P8_array)  
Q1_array = np.array(Q1_array)  
Q2_array = np.array(Q2_array)  
Q3_array = np.array(Q3_array)  
Q8_array = np.array(Q8_array)  

Frequencies = pd.read_csv(folder+"Frequencies.csv",index_col = False,header=None)
Voltages = pd.read_csv(folder+"Voltages_load.csv",index_col = False,header=None)
Voltages_gen = pd.read_csv(folder+"Voltages_gen.csv",index_col = False,header=None)
Active_power = pd.read_csv(folder+"Active_power.csv",index_col = False,header=None)
Reactive_power = pd.read_csv(folder+"Reactive_power.csv",index_col = False,header=None)
N = 370
X = np.zeros((N,20))
Y = np.zeros((N,5))
for k in range(N):
    x1 = [P1_array[k+1][1],P2_array[k+1][1],P3_array[k+1][1],P8_array[k+1][1],Q1_array[k+1][1],Q2_array[k+1][1],Q3_array[k+1][1],Q8_array[k+1][1]]
    x2 = Voltages_gen.iloc[50+k].tolist()[1:]
    x3 = Active_power.iloc[50+k].tolist()[1:]
    x4 = Reactive_power.iloc[50+k].tolist()[1:]
    X[k] = np.array([c for c in x1]+[c for c in x2]+[c for c in x3]+[c for c in x4])
    Y[k] = np.array([x-y for x,y in zip(Voltages.iloc[51+k].tolist()[1:],Voltages.iloc[50+k].tolist()[1:])]+[sum(Frequencies.iloc[51+k].tolist()[1:])/4.0-sum(Frequencies.iloc[50+k].tolist()[1:])/4.0])
W = (np.dot(np.dot(np.linalg.pinv(np.dot(X.T,X)),X.T),Y))
Y_pred = np.dot(X,W)
tr_err = 0
for k in range(N):
    for i in range(5):
        tr_err += abs(Y_pred[k,i]-Y[k,i])/5/N

print(f"Training error: {tr_err}")

N_test = 30
X_test = np.zeros((N_test,20))
Y_test = np.zeros((N_test,5))
for k in range(N_test):
    x1 = [P1_array[N+k+1][1],P2_array[N+k+1][1],P3_array[N+k+1][1],P8_array[N+k+1][1],Q1_array[N+k+1][1],Q2_array[N+k+1][1],Q3_array[N+k+1][1],Q8_array[N+k+1][1]]
    x2 = Voltages_gen.iloc[50+N+k].tolist()[1:]
    x3 = Active_power.iloc[50+N+k].tolist()[1:]
    x4 = Reactive_power.iloc[50+N+k].tolist()[1:]
    X_test[k] = np.array([c for c in x1]+[c for c in x2]+[c for c in x3]+[c for c in x4])
    Y_test[k] = np.array([x-y for x,y in zip(Voltages.iloc[51+N+k].tolist()[1:],Voltages.iloc[50+N+k].tolist()[1:])]+[sum(Frequencies.iloc[51+N+k].tolist()[1:])/4.0-sum(Frequencies.iloc[50+N+k].tolist()[1:])/4.0])

Y_pred = np.dot(X_test,W)
ts_err = 0
for k in range(N_test):
    for i in range(5):
        ts_err += abs(Y_pred[k,i]-Y_test[k,i])/5/N_test
print(f"Test error: {ts_err}") 

for i in range(5):
    fig, (ax) = plt.subplots(figsize=(10,5))
    t = np.arange(0, N_test, 1)
#     ax.plot(t,Y_test[:,i],t,np.dot(X_test,W)[:,i],'-','dotted')
    ax.plot(t,Y_test[:,i],t,np.dot(X_test,W)[:,i],'-')
    
    fig, (ax) = plt.subplots(figsize=(10,5))
    t = np.arange(0, N_test, 1)
    ax.plot(t,abs(Y_test[:,i]-np.dot(X_test,W)[:,i]))
```
